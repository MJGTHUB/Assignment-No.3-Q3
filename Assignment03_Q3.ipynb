## Assume,

### H0 : All praportions are equal

### H1 : All praportions are not equal. That is Male - Femail buyer ration is not similar accors region.


```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/BuyerRatio.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
male_p= df.iloc[0,1:5]
female_p=df.iloc[1,1:5]
buyer=[[male_p],[female_p]]
```

```python
p_value = stats.chi2_contingency([[male_p],[female_p]])
```

Or

```python
p_value = stats.chi2_contingency(buyer)
```

```python
p_value
```

```python
p_value[1]
```

## Conclusion:

p_value = 0.6603094907091882 which is greater than 'alpha'. which is 0.05 (5% significant level).

so, we accept H0 : All praportions are equal

Means, Male - Femail buyer ration is not similar accors region.
